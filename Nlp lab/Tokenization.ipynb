```python
text="""varsha is my name . i mean varrrssshhhaaaa . ok"""
```

```python
w=text.split()
s=text.split(".")
print(w)
print(s)
```

```python
import re
wo=re.findall('[\w]+',text)
se=re.compile('[.]').split(text)
print(wo)
print(se)
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize,sent_tokenize
wor=word_tokenize(text)
sen=sent_tokenize(text)
print(wor)
print(sen)
```

```python
import spacy
!python -m spacy download en_core_web_sm
```

```python
nlp=spacy.load('en_core_web_sm')
doc=nlp(text)
wr=[]
sr=[]
for t in doc:
  wr.append(t.text)
for t in doc.sents:
  sr.append(t.text)
print(wr)
print(sr)
```
