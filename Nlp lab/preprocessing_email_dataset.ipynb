```python
from nltk.corpus import stopwords
sw=set(stopwords.words('english'))
sw.add('color')
sw.add('subject')
def rsw(text):
  return ' '.join([word for word in str(text).split() if word not in sw])

df['text']=df['text'].apply(lambda x:rsw(x))
```
