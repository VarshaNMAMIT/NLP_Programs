```python

def ngram(k):
    l = []
    dic = {}

    # Generate k-grams
    for i in range(len(w) - k + 1):
        l.append(w[i:i + k])  # Slice to create k-grams

    # Combine tokens in each k-gram with a space
    l = [' '.join(gram) for gram in l]

    # Initialize dictionary for counting
    for gram in l:
        dic[gram] = 0

    # Count occurrences of each k-gram
    for gram in l:
        dic[gram] += 1

    return dic

# Usage example
k = 3
print(ngram(k))
```
