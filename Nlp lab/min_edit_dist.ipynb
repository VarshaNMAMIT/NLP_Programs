```python
def med(s1,s2,m,n):
  me=[[0 for i in range(n+1)]for i in range(m+1)]
  for i in range(m+1):
    for j in range(n+1):
      if i==0:
        me[i][j]=j
      elif j==0:
        me[i][j]=i
      elif s1[i-1]==s2[j-1]:
        me[i][j]=me[i-1][j-1]
      else:
        me[i][j]=1+min(me[i-1][j-1],me[i][j-1],me[i-1][j])
  return me[m][n]
```

```python
if __name__=="__main__":
  str1="swathi"
  str2="varsha"
  print("minimum edit distance between 2 strings :",med(str1,str2,len(str1),len(str2)))
```
